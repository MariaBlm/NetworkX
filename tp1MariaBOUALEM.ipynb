## Réseau pondéré à partir d'un fichier txt

```python
G_df = pd.read_csv("/villes.txt", delim_whitespace=True, 
                    names=['Ville1', 'Ville2', 'weight'])
G_df
```

```python
G= nx.from_pandas_edgelist(G_df, 
                            source='Ville1',
                            target='Ville2',
                            edge_attr='weight',
                            create_using=nx.Graph())
```

```python
for i in range(0,len(G_df)):
    print(G_df.iloc[i]["Ville1"])
    print(G[G_df.iloc[i]["Ville1"]][G_df.iloc[i]["Ville2"]])
```

## Pour chaque ville associe un attribut ‘population’ qui contient sa population

```python
G3_df = pd.read_csv("/populations.txt", 
                   delim_whitespace=True, names=['Ville', 'population'])
G3_df
```

```python
G3_df.drop(['Ville'], axis=1, inplace = True)
G3_df.dropna(inplace=True)
#remove "," in population values
for index, item in enumerate(G3_df['population']):
    G3_df.iloc[index]['population'] = item.split(',')[0]
#remove '' in the index 
G3_df.index = G3_df.index.str[1:-1]
#change the  population values type  from string into integer
G3_df['population'] = G3_df['population'].astype('int64')

G3_df
```

```python
#associate  attribute ‘population’ for  ville 
for i in list(G.nodes()):
    G.nodes[i]['population'] = G3_df.loc[i].population
```

## Afficher la structure du réseau

```python
fig, ax = plt.subplots(figsize=(13,10))
pos=nx.circular_layout(G) 
nx.draw_networkx(G,pos)
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G,pos, edge_labels=labels )
plt.show()
```

## Afficher une version améliorer du réseau en utilisant les contraintes

```python
node_size = [0.0003 * nx.get_node_attributes(G, 'population')[v] for v in G]
node_color = [G.degree(v) for v in G]
edge_width = [0.003 * G[u][v]['weight'] for u, v in G.edges()]

plt.figure(figsize =(17, 14))
pos = nx.circular_layout(G)

# circular layout
nx.draw_networkx(G, pos, node_size = node_size,font_color='#DC143C',
                node_color =node_color ,
                alpha = 1,
                with_labels = True, 
                width = edge_width,
                font_size=13,
                edge_color ='#68838B', cmap = plt.cm.Blues,font_family='sans-serif' )
```
